```python
import pandas as pd
import matplotlib.pyplot as plt

leitor_pda = pd.read_csv('PDA_PNME_2017_2018_2019.csv', sep= ';', encoding='latin-1')

print(leitor_pda.head()) 
print(leitor_pda.columns)

o_municipio = leitor_pda['NOME_MUNICIPIO'].value_counts().nlargest(25)

print(o_municipio)
o_municipio.plot(kind='bar')
plt.xlabel('NOME MUNICÍPIO')



```

```python
o_alunado = leitor_pda['QUANT_ALUNADO'].value_counts().nlargest(10)

print(o_alunado)
o_alunado.plot(kind='pie', autopct='%1.1f%%')
plt.title('QUANTIDADE ALUNADO')
plt.show()
```
